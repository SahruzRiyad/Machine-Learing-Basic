```python
import numpy as np
from matplotlib import pyplot as plt

tf = np.array([True , False])
dataset = np.empty((1000,4) , dtype ='O')
dataset.shape
```

```python
dataset[:,:1] = np.random.randint(0,30,1000).reshape(1000,1)
dataset[:,1:2] = np.random.choice(tf,1000).reshape(1000,1)
dataset[:,2:3] = np.random.choice(tf,1000).reshape(1000,1)
max_salary = 1000000
for i in range(0,1000):
    dataset[i][3] = (max_salary + dataset[i][1] * 1000 + dataset[i][2] * 2000) / (31-dataset[i][0])
dataset
```

```python
dataset.columns = ['Experience' , 'Front_End' , 'Back_End' , 'Salary']
dataset
```

```python
plt.plot(dataset.Experience)
plt.show()
```

```python
plt.plot(dataset.Front_End)
plt.show()
```

```python
plt.plot(dataset.Back_End)
plt.show()
```

```python
plt.scatter(dataset.Experience,dataset.Salary , marker = '*' , color = 'red')
plt.xlabel('Experience');
plt.ylabel('Salary')
```

```python
plt.scatter(dataset.Front_End,dataset.Salary , marker = '.' , color = 'blue')
plt.xlabel('Front_End');
plt.ylabel('Salary')
```

```python
plt.scatter(dataset.Back_End,dataset.Salary , marker = 'o' , color = 'green')
plt.xlabel('Back_End');
plt.ylabel('Salary')
```
